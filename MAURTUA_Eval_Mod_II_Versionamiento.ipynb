## **a.**

Visualiza la función $seno$ junto con varias rectas tangentes en diferentes puntos (`0, np.pi/4, np.pi/2, np.pi, 3*np.pi/2`). El resultado debe ser practicamente al que a continuacion muestro.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Dominio y función
dominio = np.linspace(-0.5, 2*np.pi, 500)
funcion = np.sin(dominio)

# Valores donde se evalúan las tangentes
x_tangencia = [0, np.pi/4, np.pi/2, np.pi, 3*np.pi/2]
labels_math = ['0', 'π/4', 'π/2', 'π', '3π/2']

colores_linea = ["red", "green", "blue", "magenta", "cyan"]

# Figura
plt.figure(figsize=(12, 6))

plt.plot(
    dominio,
    funcion,
    color="black",
    linewidth=2.8,
    label=r"$f(x)=\sin(x)$"
)

# Tangentes y puntos
for x0, etiqueta, col in zip(x_tangencia, labels_math, colores_linea):
    y0 = np.sin(x0)
    pendiente = np.cos(x0)

    x_local = np.linspace(x0 - 1, x0 + 1, 15)
    y_local = pendiente * (x_local - x0) + y0

    plt.plot(
        x_local,
        y_local,
        color=col,
        linewidth=2.3,
        label=f"Tangente en {etiqueta}"
    )

    plt.scatter(
        x0,
        y0,
        color=col,
        s=70,
        zorder=6
    )

# Estética del gráfico
plt.title("Rectas tangentes a f(x) = sin(x)", fontweight="bold")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="gray", linewidth=0.6, alpha=0.6)
plt.grid(True, alpha=0.3)
plt.xlim(-0.5, 2*np.pi)
plt.ylim(-1.25, 1.5)
plt.legend(loc="upper right", fontsize="small")

plt
```

```python
# Guardar imagen
ruta = "/content/cap199_preg2_mod2"
os.makedirs(ruta, exist_ok=True)

archivo = f"{ruta}/tangentes_seno.png"
plt.savefig(archivo, dpi=300, bbox_inches="tight")
plt.show()

archivo
```
